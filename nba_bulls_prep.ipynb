<h4>Notebook for Basketball Data Preparation</h4>
<p>We will take a closer look at the game schedules of Chicago's NBA team, the Chicago Bulls, of end of season 2021 and beginning of season 2022. The 2021 season data can be found here: https://fixturedownload.com/results/nba-2021/chicago-bulls and the 2022 season data here: https://fixturedownload.com/results/nba-2022/chicago-bulls</p>

```python
# Importing the libraries
import pandas as pd
```

```python
# Importing Chicago Bears 2021 season's schedule
IMPORT_FILE_PATH = "./data/input/"
nba_cb_2021 = pd.read_csv(f"{IMPORT_FILE_PATH}nba-2021-chicago-bulls-UTC.csv", parse_dates=["Date"], dayfirst=True)
```

```python
# Inspecting the dataset
nba_cb_2021.info()
nba_cb_2021.head(5)
```

```python
# Selecting rows from year 2022 only and saving as new dataframe
nba_cb_2021_22 = nba_cb_2021[(nba_cb_2021["Date"].dt.year) == 2022]

nba_cb_2021_22.info()
nba_cb_2021_22.head(15)
```

```python
# Importing Chicago Bears 2022 season's schedule
nba_cb_2022 = pd.read_csv(f"{IMPORT_FILE_PATH}nba-2022-chicago-bulls-UTC.csv", parse_dates=["Date"], dayfirst=True)
```

```python
# Inspecting the 2022 dataset
nba_cb_2022.info()
nba_cb_2022.head(5)
```

```python
# Selecting rows from year 2022 only and saving as new dataframe
nba_cb_2022_22 = nba_cb_2022[(nba_cb_2022["Date"].dt.year) == 2022]

nba_cb_2022_22.info()
nba_cb_2022_22.head(15)
```

```python
# Combining the season 21 and 22 data
nba_bulls_2022 = pd.concat([nba_cb_2021_22, nba_cb_2022_22], axis=0, ignore_index=True, sort=False)

nba_bulls_2022
```

```python
# Deleting unnecessary columns "Match Number", "Round Number", "Result"
nba_bulls_red = nba_bulls_2022.drop(["Match Number", "Round Number", "Result"], axis=1)

nba_bulls_red
```

```python
# Select rows where Chicago Bulls were the home team and drop all other rows
nba_bulls_home = nba_bulls_red[nba_bulls_red["Home Team"] == "Chicago Bulls"]
nba_bulls_home
```

```python
# Splitting column "Date" into two separate columns "Date" and "Time"

nba_bulls_home['Time'] = pd.to_datetime(nba_bulls_home['Date']).dt.time
nba_bulls_home['Date'] = pd.to_datetime(nba_bulls_home['Date']).dt.date
```

```python
nba_bulls_home
```

```python
# Exporting the Chicago Bulls 2022 matches to a csv-file
OUTPUT_FILE_PATH = "./data/prepared/"
nba_bulls_home.to_csv(f"{OUTPUT_FILE_PATH}nba_bulls_2022.csv", index=False)
```
