## Imports

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from datetime import datetime, timedelta, date, time
import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)
```

Import CSV file

```python
Taxi_data = pd.read_csv('C:/Users/OGSKNECH/AAA/aaa/data/prepared/taxi_trips.csv')
Taxi_data.head(10)
```

```python
# Convert the start time to the date format
Taxi_data['start_time'] = pd.to_datetime(Taxi_data['start_time'])

# Adding columns for analysis
Taxi_data['Time'] = Taxi_data['start_time'].dt.time
Taxi_data['Hour'] = Taxi_data['start_time'].dt.hour
Taxi_data['Month'] = Taxi_data['start_time'].dt.month
Taxi_data['Weekday'] = Taxi_data['start_time'].dt.dayofweek
Taxi_data['Day'] = Taxi_data['start_time'].dt.date
Taxi_data['Day'] = pd.to_datetime(Taxi_data['Day'])
```

## Time Demand

```python
# Average cab rides per hour as bar chart
plt.style.use("fivethirtyeight")
plt.figure(figsize=(15, 8))
taxi_by_hours = Taxi_data.groupby('Hour')['trip_id'].count().div(365)
ax = taxi_by_hours.plot(title="Average cab rides per hour as bar chart", legend=False, kind='bar')
ax.set_xlabel('Hour')
ax.set_ylabel('Trips')
```

Overview of the trips every hour, it can be seen that between 23.00 - 05.00 the demand is low and then the demand increases sharply.

```python
# Average cab trips per hour of the day
plt.style.use("fivethirtyeight")
plt.figure(figsize=(15, 5))
avg_trips_per_hour = Taxi_data.groupby(['Day', 'Hour'])['trip_id'].count().groupby('Hour').mean()
ax = avg_trips_per_hour.plot(title="Average trips per hour of the day")
ax.set_xlabel('Hour')
ax.set_ylabel('Average trips')
```

```python
#Count Taxi trips by Day of whole year
plt.style.use("fivethirtyeight")
 
# Setting figure size to 15, 10
plt.figure(figsize=(50, 30))

bikes_by_day = Taxi_data.pivot_table("trip_id", aggfunc='count', index="Day" )
ax = bikes_by_day.plot(title ="Daily amount of taxi trips", legend=False)
ax.set_xlabel("Day")
ax.set_ylabel("Taxi trips")
```

```python
#Taxi Trips per month
monthly_trips = Taxi_data.pivot_table('trip_id', aggfunc='count', index='Month')
monthly_trips = monthly_trips.rename(index={1: 'Jan', 2: 'Feb', 3: 'MÃ¤r', 4: 'Apr', 5: 'Mai', 6: 'Jun',
                                            7: 'Jul', 8: 'Aug', 9: 'Sep', 10: 'Okt', 11: 'Nov', 12: 'Dez'})
plt.style.use("fivethirtyeight")
plt.figure(figsize=(15, 15))
ax = monthly_trips.plot(title="Taxi trips per Month", legend=False, kind='bar')
ax.set_ylabel('Trips')
```

```python
# setting figure size to 15, 15
plt.figure(figsize=(40, 40))

#Average Taxi Trips per Season 
# Labelling the axes and setting
x = Taxi_data.copy()
x["season"] = x.Day.dt.month%12 // 3 + 1
season_time = x.pivot_table("trip_id", aggfunc ='count', index = "season" )
season_time = season_time.rename(index={ 1: "winter",2: "spring", 3: "summer", 4:"autumn"})
ax = season_time.plot(title ="Average Taxi trips per season", legend=False, kind ="bar")
ax.set_ylabel("Taxi Trips")
ax.set_xlabel("Seasons")

plt.gcf().axes[0].yaxis.get_major_formatter().set_scientific(False)
```

```python
#Plot average Taxi Trips per Hour and Split between Weekdays and Weekends
hour = Taxi_data.pivot_table('trip_id', aggfunc='count',index=['Hour','Day']).fillna(0).reset_index('Hour')

hour['weekend'] = (hour.index.dayofweek >= 5)
hour = hour.groupby(['weekend', 'Hour']).mean()
hour.index.set_levels([['weekday', 'weekend'],
                          ["{0}:00".format(x) for x in range(24)]], inplace=True);
hour.columns.name = None

#Create Subplots as bars
fig, ax = plt.subplots(1, 2, figsize=(20, 10), sharey=True)
hour.loc['weekday'].plot(title='Taxi Trips on Weekdays (Mo-Fr)', ax=ax[0], kind = 'bar')
hour.loc['weekend'].plot(title='Taxi Trips on Weekends (Sa-Su)', ax=ax[1], kind = 'bar')
ax[0].set_ylabel('Average Taxi Trips per Hour')
```

## Price Demand

```python
# Count number of payments by payment type
payment_counts = Taxi_data['payment_type'].value_counts()
print(payment_counts)

# Visualization with a bar chart
plt.figure(figsize=(10, 6))
payment_counts.plot(kind='bar')
plt.xlabel('Paymenttype')
plt.ylabel('Number of payments')
plt.title('Number of payments by payment type')

plt.show()
```

```python
# Calculate average tips per hour
Taxi_data['hour'] = Taxi_data['start_time'].dt.hour
tip_per_hour = Taxi_data.groupby('hour')['tips'].mean().reset_index()

plt.figure(figsize=(10, 6))
sns.barplot(x='hour', y='tips', data=tip_per_hour)
plt.xlabel('Hour')
plt.ylabel('Average tips')
plt.title('Average tips per hour')
plt.ylim(0, 4)
plt.show()
```

```python
Taxi_data['total'] = Taxi_data['total'].str.replace('$', '').str.replace(',', '').astype(float)
Average_price = Taxi_data['total'].mean()
print('Average price of the trips:', Average_price)

# Group the trips by month and calculate the average price
Average_price = Taxi_data.groupby('Month')['total'].mean()

# Create a bar chart to visualize the average price by month
plt.figure(figsize=(10, 6))
average_price.plot(kind='bar')
plt.xlabel('Month')
plt.ylabel('Average price')
plt.title('Average price by months')
plt.show()
```

```python
# Convert the 'total' column to numeric type
Taxi_data['total'] = pd.to_numeric(Taxi_data['total'], errors='coerce')

# Create a histogram to visualize the distribution of the total prices
plt.figure(figsize=(10, 6))
plt.hist(Taxi_data['total'], bins=10000, range=(0,100), edgecolor='k')
plt.xlabel('Total price')
plt.ylabel('Amount')
plt.title('Distribution of total prices')
plt.ylim(0, 200000)
plt.show()
```

```python
# Group the data set by the hour of the day and calculate the average price
average_prices_by_hour = Taxi_data.groupby(Taxi_data['start_time'].dt.hour)['total'].mean()

# Create the line chart to visualize the distribution of prices by hours in the day
plt.figure(figsize=(10, 6))
plt.plot(average_prices_by_hour.index, average_prices_by_hour.values)
plt.xlabel('Hour of the day')
plt.ylabel('Average price')
plt.title('Distribution of prices by hours per day')
plt.ylim(15, 40)

plt.show()
```
