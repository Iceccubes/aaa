```python
# Import Weather Datset
df = pd.read_csv('taxi_weather.csv')
df
```

```python
# Importing LabelEncoder from Sklearn
# library from preprocessing Module.
from sklearn.preprocessing import LabelEncoder
 
# Creating a instance of label Encoder.
le = LabelEncoder()
 
#payment_type'     company  condition
# Using .fit_transform function to fit label
# encoder and return encoded label
#label = le.fit_transform(df['condition'])
label = le.fit_transform(df['company'])

# printing label
label
# removing the column 'Purchased' from df
# as it is of no use now.
df.drop("company", axis=1, inplace=True)
 
# Appending the array to our dataFrame
# with column name 'Purchased'
df["company"] = label
 
# printing Dataframe
df
```

```python
df=df.drop(columns='start_time')
df=df.drop(columns='end_time')
df=df.drop(columns='trip_id')
df=df.drop(columns='taxi_id')
df=df.drop(columns='pickup_census_tract')
df=df.drop(columns='dropoff_census_tract')

df=df.drop(columns='pickup_centroid_location')
df=df.drop(columns='dropoff_centroid_location')

df
#df.to_csv('data.csv', index=False)
```

```python
df.fillna(0)
df.replace(np.nan, 0)
df
```

```python
df['pickup_community_area'] = df['pickup_community_area'].replace(np.nan, 0)
df['dropoff_community_area'] = df['dropoff_community_area'].replace(np.nan, 0)
df['pickup_centroid_latitude'] = df['pickup_centroid_latitude'].replace(np.nan, 0)
df['pickup_centroid_longitude'] = df['pickup_centroid_longitude'].replace(np.nan, 0)
df['dropoff_centroid_latitude'] = df['dropoff_centroid_latitude'].replace(np.nan, 0)
df['dropoff_centroid_longitude'] = df['dropoff_centroid_longitude'].replace(np.nan, 0)
print (df.dtypes)
df.isnull().sum()
```

```python
from sklearn.cluster import KMeans
km = KMeans(n_clusters=19).fit(df)
# 建立模型，K=19

df['cluster'] = km.labels_
#df.sort_values('cluster')
df.to_csv('cluster.csv', index=False)
```

```python
import numpy as np
from sklearn.mixture import GaussianMixture
df=df.drop(columns='cluster')
gmm = GaussianMixture(n_components=4)
gmm.fit(df)
labels = gmm.predict(df)
df['cluster']=labels 
print(labels)
df.to_csv('cluster_gmm.csv', index=False)
```
