```python
#import the prepared datasets
taxi_trips = pd.read_csv('data/prepared/taxi_trips.csv')
weather_data = pd.read_csv('data/prepared/weather_data_2022.csv')
```

### Merge Datasets

```python
#merge taxi trips on start time with weather data on date time
taxi_weather = pd.merge(taxi_trips, weather_data, left_on='start_time', right_on='date_time', how='left')
#seconds time not needed
taxi_weather = taxi_weather.drop('date_time', axis=1)
taxi_weather
```

```python
#export merged taxi-weather dataset
taxi_weather.to_csv('data/prepared/taxi_weather.csv', index=False)
```
